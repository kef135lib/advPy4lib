# Imagine generation, just for fun


This notebook will use some Forms elements and Hugging Face Libraries to create images based on text prompts. Remember to change the runtime to the GPU option and to copy to your drive before running this notebook.

```python
#Don't forget to change runtime!

!pip install transformers

from transformers import pipeline
from IPython.display import display, Markdown
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler
import torch
from google.colab import files

model_id = "sd2-community/stable-diffusion-2-1"
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

display(Markdown("## Ready to proceed"))
```

```python
# @title Text to imagine
prompt = "a noir movie poster featuring a librarian that is acting like a 1940's detective" # @param {"type":"string"}
file_name = "detective" # @param {"type":"string"}
image = pipe(prompt).images[0]
display(image)
```

```python
#Wanna download?

image.save(f"{file_name}.png")
files.download(f"{file_name}.png")
```
